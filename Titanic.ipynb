```python
import pandas as pd
df=pd.read_csv(r'D:\Old\vscodess\Python\EDA\train.csv')
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df.drop(columns=['Cabin'], inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x='Pclass', y='Survived', data=df, errorbar=None)
plt.title('Survival Rate by Passenger Class')
plt.show()

sns.barplot(x='Pclass', y='Survived', data=df, estimator=sum, errorbar=None)
plt.title('Number of Survivors by Passenger Class')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df, errorbar=None)
plt.title('Survival Rate by Gender')
plt.show()

sns.barplot(x='Sex', y='Survived', data=df, estimator=sum, errorbar=None)
plt.title('Number of Survivors by Gender')
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 18, 35, 60, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])
sns.barplot(x='AgeGroup', y='Survived', data=df, errorbar=None)
plt.title('Survival Rate by Age Group')
plt.show()

sns.barplot(x='AgeGroup', y='Survived', data=df, estimator=sum, errorbar=None)
plt.title('Number of Survivors by Age Group')
plt.show()
```

```python
df['Total_Members'] = df['SibSp'] + df['Parch']
sns.barplot(x='Total_Members', y='Survived', data=df, errorbar=None)
plt.title('Survival Rate by Size of Family')
plt.show()

sns.barplot(x='Total_Members', y='Survived', data=df, estimator=sum, errorbar=None)
plt.title('Number of Survivors by Size of Family')
plt.show()
```

```python
sns.barplot(x='Embarked', y='Survived', data=df, errorbar=None)
plt.title('Survival Rate by Embarkation Port')
plt.show()

sns.barplot(x='Embarked', y='Survived', data=df, estimator=sum, errorbar=None)
plt.title('Number of Survivors by Embarkation Port')
plt.show()
```

```python
sns.countplot(x='AgeGroup', hue='Survived', data=df)
plt.title('Number of Survivors by Age Group')
plt.show()
```
